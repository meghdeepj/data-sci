### MNIST Digit Recognition

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import torch as T
import torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
```

#### Data Pre-processing

```python
transform=transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,), (0.5,))])
```

```python
trainset= datasets.MNIST('/files/', train=True, download=True, transform=transform)
testset= datasets.MNIST('/files/', train=False, download=True, transform=transform)
```

```python
trainset, valset= T.utils.data.random_split(trainset, [50000, 10000])
```

```python
train_loader = T.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
val_loader = T.utils.data.DataLoader(valset, batch_size=10000, shuffle=True)
test_loader = T.utils.data.DataLoader(testset, batch_size=10000, shuffle=True)
```

```python
idx, (imgs, labels) = next(enumerate(test_loader))
```

```python
fig = plt.figure()
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.tight_layout()
    plt.imshow(imgs[i][0], cmap='gray', interpolation='none')
    plt.title("Ground Truth: {}".format(labels[i]))
    plt.xticks([])
    plt.yticks([])
```

```python
class JanNet(nn.Module):
    def __init__(self):
        super(JanNet, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=5, stride=1, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))
        self.layer2 = nn.Sequential(
            nn.Conv2d(16, 32, kernel_size=5, stride=1, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))
        self.fc1 = nn.Linear(7*7*32, 10)
        self.fc3 = nn.LogSoftmax(dim=1)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.view( -1, 7*7*32)
        out = self.fc1(out)
        out = self.fc3(out)
        return out
janNet=JanNet().to('cuda')
```

```python
print(janNet)
```

```python
criterion=nn.NLLLoss()
optimizer=optim.Adam(janNet.parameters(), lr=0.002, betas=(0.9,0.999))
```

```python
n_epochs=5
train_loss=[]
train_count=[]
counter=0
val_counter=0
val_loss=[]
val_count=[]

for epoch in range(n_epochs):
    janNet.train()
    for idx, (img, labels) in enumerate(train_loader):
        img = img.to('cuda')
        labels= labels.to('cuda')
        print(img.shape, labels.shape)
        optimizer.zero_grad()
        output=janNet(img)
        print(output[0], labels[0])
        loss=criterion(output,labels)
        loss.backward()
        optimizer.step()
        train_loss.append(loss.item())
        train_count.append(counter)
        counter+=1
    
    janNet.eval()
    for idx, (img, labels) in enumerate(val_loader):
        img=img.to('cuda')
        labels=labels.to('cuda')
        with T.no_grad():
            out=janNet(img)
            valLoss=criterion(out,labels)
    val_loss.append(valLoss.item())
    val_count.append(counter)
#     val_counter+=1
    print('Epoch {} - Training Loss: {} | Val Loss: {}'.format(epoch, loss.item(), valLoss.item()))
T.save(janNet.state_dict(), 'model.pth')
T.save(optimizer.state_dict(), 'optimizer.pth')
```

```python
print(val_loss)
print(val_count)
```

```python
fig = plt.figure()
plt.plot(train_count, train_loss, color='blue')
plt.scatter(val_count, val_loss, color='red')
plt.legend(['Train Loss', 'Val Loss'], loc='upper right')
plt.xlabel('number of training examples seen')
plt.ylabel('negative log likelihood loss')
fig
```

```python

for idx, (img, labels) in enumerate(test_loader):
        img=img.to('cuda')
        labels=labels.to('cuda')
        with T.no_grad():
            out=janNet(img)
            valLoss=criterion(out,labels)
    val_loss.append(valLoss.item())
    val_count.append(counter)
```
