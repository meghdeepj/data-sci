```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import tensorflow as tf
import tensorflow.keras as K
```

```python
from tensorflow.keras.layers import Dense, Flatten, MaxPooling2D, Conv2D, Dropout, Activation, Softmax
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.optimizers import RMSprop, Adam
```

```python
(x_tr,y_tr),(x_te,y_te) = cifar10.load_data()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_te, x_val, y_te, y_val = train_test_split(x_te,y_te, test_size=0.2)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_tr_oh = to_categorical(y_tr, num_classes=10)
y_te_oh = to_categorical(y_te, num_classes=10)
y_val_oh = to_categorical(y_val, num_classes=10)
```

```python
x_tr=x_tr/255
x_te=x_te/255
x_val=x_val/255
```

```python
model=K.Sequential()

model.add(Conv2D(32,(3,3),padding='same',activation='relu',input_shape=(32,32,3)))
model.add(Conv2D(64,(3,3),padding='same',activation='relu',input_shape=(32,32,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(64,(3,3),padding='same',activation='relu',input_shape=(32,32,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
opt=Adam(lr=0.0005, decay=1e-7)
```

```python
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(x_tr, y_tr_oh, batch_size=128, epochs=20, verbose=1, validation_data=(x_val,y_val_oh))
```

```python
score=model.evaluate(x_te,y_te_oh,batch_size=128)
```

```python
pred=model.predict(x_te, batch_size=128)
```

```python
pred_clas=model.predict_classes(x_te, batch_size=128).reshape(-1,1)
```

```python
import numpy as np
pred_classes=np.argmax(pred, axis=1).reshape(-1,1)
```

```python
print(pred_classes)
```

```python
print(pred_clas)
```

```python
total=(pred_clas==y_te).sum()
```

```python
te_acc=total*100/y_te.shape[0]
```

```python
print('testing accuracy= ',te_acc)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()
```

```python
model.save('model_cifar.h5')
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='lower left')
plt.show()
```

```python
model1=K.Sequential()

model1.add(Conv2D(32,(3,3),padding='same',activation='relu',input_shape=(32,32,3)))
model1.add(Conv2D(64,(3,3),padding='same',activation='relu',input_shape=(32,32,3)))
model1.add(MaxPooling2D(pool_size=(2,2)))
model1.add(Dropout(0.25))
model1.add(Conv2D(64,(3,3),padding='same',activation='relu',input_shape=(32,32,3)))
model1.add(MaxPooling2D(pool_size=(2,2)))
model1.add(Dropout(0.25))
model1.add(Flatten())
model1.add(Dense(512, activation='relu'))
model1.add(Dropout(0.4))
model1.add(Dense(128, activation='relu'))
model1.add(Dense(10, activation='softmax'))
```

```python
model1.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen=ImageDataGenerator(
        rotation_range=10,
        width_shift_range=0.1,
        height_shift_range=0.1,
        zoom_range=0.05,
)
```

```python
datagen.fit(x_tr)
```

```python
model1.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history1=model1.fit_generator(datagen.flow(x_tr,y_tr_oh, batch_size=128), epochs=200, steps_per_epoch=x_tr.shape[0]//128, validation_data=(x_val,y_val_oh))
```

```python
import numpy as np
pred=model1.predict(x_te, batch_size=128)
pred_clas=model1.predict_classes(x_te, batch_size=128).reshape(-1,1)
pred_classes=np.argmax(pred, axis=1).reshape(-1,1)
```

```python
total=(pred_clas==y_te).sum()
te_acc=total*100/y_te.shape[0]
print('testing accuracy= ',te_acc)
```

```python
import matplotlib.pyplot as plt
plt.plot(history1.history['accuracy'], label='accuracy')
plt.plot(history1.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot(history1.history['loss'], label='loss')
plt.plot(history1.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='lower left')
plt.show()
```
