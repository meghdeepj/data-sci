```python
data_clean=data_clean.drop(['Price'], axis=1)
data_clean.describe(include='all')
```

```python
sns.scatterplot(x=data_clean.Year, y= data_clean.Mileage)
```

```python
# do PCA on years and mileage, combine to one variable instead of dropping

data_no_cor = data_clean.drop(['Year'], axis=1)
data_no_cor.describe(include='all')
```

```python
data_dum = pd.get_dummies(data_no_cor)
data_dum.describe(include='all')
```

```python
data_dum.columns.values
```

```python
col = ['log_price', 'Mileage', 'EngineV', 'Brand_Audi', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_crossover', 'Body_hatch',
       'Body_other', 'Body_sedan', 'Body_vagon', 'Body_van',
       'Engine Type_Diesel', 'Engine Type_Gas', 'Engine Type_Other',
       'Engine Type_Petrol', 'Registration_no', 'Registration_yes']
```

```python
data_clean = data_dum[col]
```

```python
data_clean.describe(include='all')
```

```python
target = data_clean.log_price
features =data_clean.drop(['log_price'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scl=StandardScaler()
fea_scale= scl.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(fea_scale, target, test_size=0.25, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Linear regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
p_train = reg.predict(x_train)
p_test = reg.predict(x_test)
```

```python
reg.score(x_test, y_test)
```

```python
plt.scatter(y_train, p_train)
plt.show()
plt.scatter(y_test, p_test, alpha=0.5)
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
prices = pd.DataFrame(np.exp(p_test), columns=['Prediction'])
y_test = y_test.reset_index(drop=True)
prices['Target'] = np.exp(y_test)
prices
```

```python
prices['Residual'] = prices['Target'] - prices['Prediction']
prices
```

```python
prices.Residual.mean()
```
